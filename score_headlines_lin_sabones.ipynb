```python
import joblib
from sentence_transformers import SentenceTransformer
from datetime import datetime

# Upload these via the sidebar:
# - svm.joblib
# - headlines_nyt_2024-12-02.txt

# Load headlines
input_file = "headlines_nyt_2024-12-02.txt"
source = "nyt"

with open(input_file, "r", encoding="utf-8") as f:
    headlines = [line.strip() for line in f if line.strip()]

print(f"Loaded {len(headlines)} headlines.")

# Load models
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
clf = joblib.load("svm.joblib")

# Vectorize and predict
embeddings = model.encode(headlines)
predictions = clf.predict(embeddings)

# Write output
today = datetime.now().strftime("%Y_%m_%d")
output_file = f"headline_scores_{source}_{today}.txt"

with open(output_file, "w", encoding="utf-8") as f:
    for label, headline in zip(predictions, headlines):
        f.write(f"{label}, {headline}\n")

print(f"Done! Output saved to {output_file}")
```
